```python
n_values = [10, 50, 100, 500]
sample_means = []
for n in n_values:
    means = []
    for i in range(300):
        sample = df['charges'].sample(n)
        means.append(sample.mean())
    sample_means.append(means)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.flatten()

for i, mean in enumerate(sample_means):
    axs[i].hist(mean, bins=30)
    axs[i].set_title('n = {}'.format(n_values[i]))

plt.tight_layout()
plt.show()
```

```python
for i, mean in enumerate(sample_means):
    std = np.std(mean)
    mean = np.mean(mean)
    print('n = {}, mean = {:.2f}, std = {:.2f}'.format(n_values[i], mean, std))
```

```python
mean_charges = df['charges'].mean()
std_charges = df['charges'].std()
mean_bmi = df['bmi'].mean()
std_bmi = df['bmi'].std()

print('Mean charges:', mean_charges)
print('Std charges:', std_charges)
print('Mean BMI:', mean_bmi)
print('Std BMI:', std_bmi)
```

```python
from scipy.stats import norm, t

# 95% confidence interval for charges
n = len(df)
alpha = 0.05
z = norm.ppf(1 - alpha/2)
ci_charges_95 = (mean_charges - z * std_charges / np.sqrt(n), mean_charges + z * std_charges / np.sqrt(n))

# 99% confidence interval for charges
alpha = 0.01
z = norm.ppf(1 - alpha/2)
ci_charges_99 = (mean_charges - z * std_charges / np.sqrt(n), mean_charges + z * std_charges / np.sqrt(n))

# 95% confidence interval for BMI
alpha = 0.05
t_value = t.ppf(1 - alpha/2, n-1)
ci_bmi_95 = (mean_bmi - t_value * std_bmi / np.sqrt(n), mean_bmi + t_value * std_bmi / np.sqrt(n))

# 99% confidence interval for BMI
alpha = 0.01
t_value = t.ppf(1 - alpha/2, n-1)
ci_bmi_99 = (mean_bmi - t_value * std_bmi / np.sqrt(n), mean_bmi + t_value * std_bmi / np.sqrt(n))

print('95% CI for charges:', ci_charges_95)
print('99% CI for charges:', ci_charges_99)
print('95% CI for BMI:', ci_bmi_95)
print('99% CI for BMI:', ci_bmi_99)
```

```python
from scipy.stats import probplot

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))

# q-q plot for charges
probplot(df['charges'], dist='norm', plot=axs[0])
axs[0].set_title('Q-Q plot for charges')

# q-q plot for BMI
probplot(df['bmi'], dist='norm', plot=axs[1])
axs[1].set_title('Q-Q plot for BMI')

plt.show()

from scipy.stats import kstest

# KS-test for charges
stat, p = kstest(df['charges'], 'norm', args=(mean_charges, std_charges))
print('KS-test for charges:')
print('Stat:', stat)
print('P-value:', p)

# KS-test for BMI
stat, p = kstest(df['bmi'], 'norm', args=(mean_bmi, std_bmi))
print('KS-test for BMI:')
print('Stat:', stat)
print('P-value:', p)
```

```python
import pandas as pd
from scipy.stats import shapiro, bartlett, ttest_ind

# загрузка данных
data = pd.read_csv("bmi.csv")

# выборки для сравнения
nw_bmi = data[data.region == "northwest"].bmi
sw_bmi = data[data.region == "southwest"].bmi

# проверка на нормальность
stat_nw, p_nw = shapiro(nw_bmi)
stat_sw, p_sw = shapiro(sw_bmi)

print("Шапиро-Уилк для выборки northwest: статистика=%.3f, p=%.3f" % (stat_nw, p_nw))
print("Шапиро-Уилк для выборки southwest: статистика=%.3f, p=%.3f" % (stat_sw, p_sw))

# проверка на гомогенность дисперсии
stat, p = bartlett(nw_bmi, sw_bmi)
print("Критерий Бартлетта: статистика=%.3f, p=%.3f" % (stat, p))

# сравнение средних значений выборок
t_stat, p_val = ttest_ind(nw_bmi, sw_bmi)
print("t-критерий Стьюдента: статистика=%.3f, p=%.3f" % (t_stat, p_val))
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.DataFrame({'Женат': [89,17,11,43,22,1],
                     'Гражданский брак': [80,22,20,35,6,4],
                     'Не состоит в отношениях': [35,44,35,6,8,22]})
data.index = ['Полный рабочий день','Частичная занятость','Временно не работает','На домохозяйстве','На пенсии','Учёба']

stat, p, dof, expected = chi2_contingency(data)

print("Хи-квадрат: статистика=%.3f, p=%.3f, dof=%.3f" % (stat, p, dof))
```
