```python
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from random import randrange
```

```python
digits = load_digits()

X, y = load_digits(return_X_y=True)

print('Shape of digit images: ', digits.images.shape)
print('Shape of X (training data): ', X.shape)
print('Shape of y (true labels): ', y.shape)
```

```python
fig, axs = plt.subplots(2, 5, sharey=False, tight_layout=True, figsize=(12,6), facecolor='white')
# plt.gray()
for i in range(0,2):
  for j in range(0,5):
    n = randrange(len(y))
    axs[i,j].matshow(digits.images[n])
    axs[i,j].set(title=y[n])
plt.show()
```

```python
plt.gray()
n = randrange(len(y))
plt.matshow(digits.images[n])
plt.title(y[n])
plt.show()
```

```python
%%time

# perplexities = [30]
perplexities = [5, 10, 30, 50]

for perplexity in perplexities:
  embed = TSNE(
      perplexity = perplexity,
      random_state = 42
  )

  X_embedded = embed.fit_transform(X)

  fig = px.scatter(
    x = X_embedded[:,0],
    y = X_embedded[:,1],
    color = y.astype(str),
    hover_name = y.astype(str)
  )

  fig.update_layout(
      plot_bgcolor = 'white',
      title_text = f"t-SNE (perplexity={perplexity})",
      legend_title_text = 'Numbers'
  )

  fig.update_yaxes(
      linecolor = 'black',
      gridcolor = 'lightgrey',
      zerolinecolor = 'lightgrey',
      zerolinewidth = 1
  )

  fig.update_xaxes(
      linecolor = 'black',
      gridcolor = 'lightgrey',
      zerolinecolor = 'lightgrey',
      zerolinewidth = 1
  )

  fig.show()
```

```python
%%time

for n, d in [(10,0.01), (15,0.1), (20,0.2)]:
  reducer = umap.UMAP(random_state=42, n_jobs=1, n_neighbors=n, min_dist=d)
  embedding = reducer.fit_transform(X)

  fig = px.scatter(
    x = embedding[:,0],
    y = embedding[:,1],
    color = y.astype(str),
    hover_name = y.astype(str)
  )

  fig.update_layout(
      plot_bgcolor = 'white',
      title_text = f"MNIST data embedded into two dimensions by UMAP with {n} neighbors and {d} min_distance",
      legend_title_text = 'Numbers'
  )

  fig.update_yaxes(
      linecolor = 'black',
      gridcolor = 'lightgrey',
      zerolinecolor = 'lightgrey',
      zerolinewidth = 1
  )

  fig.update_xaxes(
      linecolor = 'black',
      gridcolor = 'lightgrey',
      zerolinecolor = 'lightgrey',
      zerolinewidth = 1
  )

  fig.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE

digits = load_digits()

X, y = load_digits(return_X_y=True)

test = TSNE(
    perplexity = 30,
    random_state = 42
)

test_embeded = test.fit_transform(X)
```

```python
fig = px.scatter(
  x = test_embeded[:,0],
  y = test_embeded[:,1],
  color = y.astype(str),
  hover_name = y.astype(str),
  text = range(len(y))
)

fig.update_layout(
    plot_bgcolor = 'white',
    title_text = f"t-SNE (perplexity=30)",
    legend_title_text = 'Numbers'
)

fig.update_yaxes(
    linecolor = 'black',
    gridcolor = 'lightgrey',
    zerolinecolor = 'lightgrey',
    zerolinewidth = 1
)

fig.update_xaxes(
    linecolor = 'black',
    gridcolor = 'lightgrey',
    zerolinecolor = 'lightgrey',
    zerolinewidth = 1
)

fig.show()
```
