```python
# Task 2
import math
figures = ['triangle', 'rectangle', 'circle']
area = {}
while True:
  fig = input('Введите название фигуры из списка' f'{figures}' '\n')
  if (fig == 'exit'):
    break
  if fig == 'triangle':
    x = float(input('Введите основание треугольника: '))
    y = float(input('Введите высоту треугольника: '))
    area[fig] = x*y/2
  elif fig == 'rectangle':
    x = float(input('Введите основание прямоугольника: '))
    y = float(input('Введите высоту прямоугольника: '))
    area[fig] = x*y
  elif fig == 'circle':
    x = float(input('Введите радиус круга: '))
    area[fig] = math.pi*x**2
  else:
    print('Такой фигуры нет в списке')
  # ans = input('Хотите ввести еще площади? Y/N? \n')
  # if ans == 'n':
  #     fig = 'exit'
print(area)
```

```python
# Task 3
eq = input('Введите выражение: ')
op = ''
for symb in eq:
  if not symb.isdigit():
    op += symb
x, y = map(float, (eq.split(op)))
inst = {'+': x+y, '-': x-y, '/': x/y, '//': x//y, 'abs': f'{abs(x)}, {abs(y)}', 'pow': x**y, '**': x**y}
print(inst[op])
```

```python
# Task 4
total = 0
nums = []
while True:
  num = float(input('Введите число: '))
  total += num
  nums.append(num)
  if (total == 0):
    break
total = 0
for num in nums:
  total += num**2
print(total)
```

```python
# Task 5
n = float(input("N = "))
x = 1
nums = []
while n > 0:
  for _ in range(x):
    if n == 0:
      break
    nums.append(x)
    n -= 1
  x += 1
print(*nums)
```

```python
# Task 6
A = [ 1 ,  2 ,  3 ,  4 ,  2 ,  1 ,  3 ,  4 ,  5 ,  6 ,  5 ,  4 ,  3 ,  2]
B = ['a', 'b', 'c', 'c', 'c', 'b', 'a', 'c', 'a', 'a', 'b', 'c', 'b', 'a']
ans = {}
for id, letter in enumerate(B):
  ans[letter] = A[id] if letter not in ans else ans[letter]+A[id]
print(ans)
```
