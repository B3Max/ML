```python
# 2.2. Реализовать регрессию вручную, отобразить наклон, сдвиг и MSE.
# Manual Linear Regression Implementation
def linear_regression(x, y):
    n = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x_squared = np.sum(x ** 2)
    sum_x_times_y = np.sum(x * y)

    slope = (n * sum_x_times_y - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    intercept = (sum_y - slope * sum_x) / n

    predictions = slope * x + intercept
    mse = np.mean((predictions - y) ** 2)

    return slope, intercept, mse

# Applying linear regression to data
slope, intercept, mse = linear_regression(data['in_spotify_playlists'], data['in_apple_playlists'])

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"MSE (Mean Squared Error): {mse:.2f}")
```

```python
# 3. Загрузить данные: 'insurance.csv'.
data = pd.read_csv('insurance.csv')
data.describe()
```

```python
data.isna().sum()
```

```python
# 3. Вывести список уникальных регионов.
unique_regions = data['region'].unique()
print("Уникальные регионы:", unique_regions)
```

```python
# 3.1. Однофакторный ANOVA тест с использованием scipy
from scipy import stats

region_grouped = [data['bmi'][data['region'] == region] for region in unique_regions]
stats.f_oneway(*region_grouped)
```

```python
# 3.2. Однофакторный ANOVA тест с использованием statsmodels
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('bmi ~ region', data=data).fit()
sm.stats.anova_lm(model, typ=2)
```

```python
# 3.3. Перебор всех пар с тестом Стьюдента и поправкой Бонферрони
from itertools import combinations

pairwise_comparisons = list(combinations(unique_regions, 2))
bonferroni_correction = 0.05 / len(pairwise_comparisons)

for region1, region2 in pairwise_comparisons:
    data1 = data['bmi'][data['region'] == region1]
    data2 = data['bmi'][data['region'] == region2]
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    significant = p_value < bonferroni_correction
    print(f"T-тест между регионами {region1} и {region2}: t-статистика = {t_statistic:.2f}, p-значение = {p_value:.4f}, Значимость: {significant}")
```

```python
# 3.4. Пост-хок тест Тьюки
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=data['bmi'], groups=data['region'], alpha=0.05)
tukey.plot_simultaneous()
print(tukey.summary())
```

```python
# 3.5. Двухфакторный ANOVA тест с использованием statsmodels
two_way_model = ols('bmi ~ region * sex', data=data).fit()
sm.stats.anova_lm(two_way_model, typ=2)
```

```python
# 3.6. Пост-хок тест Тьюки для двухфакторного ANOVA
data['combination'] = data.sex + " / " + data.region
tukey_two_way = pairwise_tukeyhsd(endog=data['bmi'], groups=data['combination'], alpha=0.05)
tukey_two_way.plot_simultaneous()
print(tukey_two_way.summary())
```
